# Importing data

```python
data = ''
with open("data/shakespeare.txt", 'r') as f:
    data += f.read()
```

# Building BPE tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
from collections import defaultdict

words_freq = defaultdict(int)
words_with_offset = tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str(data)
for word, _ in words_with_offset:
    words_freq[word] += 1
```
